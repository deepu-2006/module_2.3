merge three df.
using reduse function


```python
from functools import reduce
import pandas as pd
```

```python
df1 = pd.DataFrame({"A":[1,2,3],"B":[10,20,30]})
df2 = pd.DataFrame({"A":[1,2,3],"c":[100,200,300]})
df3 = pd.DataFrame({"A":[1,2,3],"D":[1000,2000,3000]})
var1 = [df1,df2,df3]
var2 = reduce(lambda left,right:pd.merge(left,right,on="A"),var1)
var2




```

merge 3 dataframe

```python
df1 = pd.DataFrame({"A":[1,2,3],"B":[10,20,30]})
df2 = pd.DataFrame({"A":[1,2,3],"c":[100,200,300]})
df3 = pd.DataFrame({"A":[1,2,3],"D":[1000,2000,3000]})
var1 = pd.merge(df1,df2)
var2 = pd.merge(var1,df3)
var2
```

perform join on df and series.

```python
#join on df
df1 = pd.DataFrame({"A":[1,2,3,]}, index = ["a","b","c"])
df2 = pd.DataFrame({"B":[1,3]}, index = ["a","c"])
#join on index
pr1 = df1.join(df2)
pr1
```

```python
#join on series
sr1 = pd.Series([10,20,30], index = ["a","b","c"])
sr2 = pd.Series([100,200,300], index = ["a","b","c"])
pr2 = sr1.join(sr2)
pr2
```

join on concate

```python
sr1 = pd.Series([1,2,3],name = "A")
sr2 = pd.Series([4,5,6], name = "B")
sr3 = pd.Series([7,8,9],name = "C")
#concat sries one and two
result =pd.concat([sr1,sr2],axis=1)
# join series three
result.join(sr3)
```
